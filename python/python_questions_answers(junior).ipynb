# 2️⃣ Object-Oriented Programming — OOP (বস্তু-ভিত্তিক প্রোগ্রামিং)

## 🟡 সহজ ভাষায়:
**বাস্তব জগতের মতো করে চিন্তা করা।**

বাস্তবে সবকিছু **Object (বস্তু):**
- 🚗 গাড়ি → রঙ আছে, গতি আছে, চলতে পারে
- 👤 মানুষ → নাম আছে, বয়স আছে, কথা বলতে পারে
- 🏦 ব্যাংক → টাকা আছে, জমা নেওয়া যায়, তোলা যায়

OOP-তে এভাবেই **Object বানিয়ে** কোড লেখা হয়।

## 🔑 OOP-এর ৪টি মূল স্তম্ভ:

---

### 🏛️ 1. Class & Object

```python
# Class হলো Blueprint (নকশা)
# Object হলো সেই নকশা থেকে তৈরি জিনিস

class মানুষ:
    def __init__(self, নাম, বয়স):
        self.নাম = নাম
        self.বয়স = বয়স
    
    def পরিচয়(self):
        print(f"আমি {self.নাম}, আমার বয়স {self.বয়স}")

# Object তৈরি
ব্যক্তি১ = মানুষ("রহিম", 25)
ব্যক্তি২ = মানুষ("করিম", 30)

ব্যক্তি১.পরিচয়()  # আমি রহিম, আমার বয়স 25
ব্যক্তি২.পরিচয()  # আমি করিম, আমার বয়স 30
```

> 🏠 উদাহরণ: **Class = বাড়ির নকশা, Object = সেই নকশায় তৈরি বাড়ি**

---

### 🏛️ 2. Encapsulation (আবদ্ধকরণ)

**Data লুকিয়ে রাখা, বাইরে থেকে সরাসরি access না দেওয়া।**

```python
class ব্যাংক_একাউন্ট:
    def __init__(self):
        self.__balance = 0  # __ মানে Private, বাইরে দেখা যাবে না
    
    def টাকা_জমা(self, পরিমাণ):
        if পরিমাণ > 0:
            self.__balance += পরিমাণ
            print(f"{পরিমাণ} টাকা জমা হয়েছে")
    
    def balance_দেখো(self):
        return self.__balance

account = ব্যাংক_একাউন্ট()
account.টাকা_জমা(5000)
print(account.balance_দেখো())  # 5000

# সরাসরি access করা যাবে না!
# print(account.__balance)  ❌ Error আসবে
```

> 🔐 উদাহরণ: **ATM মেশিন — ভেতরের টাকা দেখা যায় না, শুধু নির্দিষ্ট উপায়ে তোলা যায়**

---

### 🏛️ 3. Inheritance (উত্তরাধিকার)

**একটি Class-এর বৈশিষ্ট্য আরেকটি Class নিয়ে নেয়।**

```python
# Parent Class
class প্রাণী:
    def __init__(self, নাম):
        self.নাম = নাম
    
    def শ্বাস_নেয়(self):
        print(f"{self.নাম} শ্বাস নিচ্ছে")

# Child Class — প্রাণীর সব বৈশিষ্ট্য পাবে + নিজেরটাও থাকবে
class কুকুর(প্রাণী):
    def ডাকে(self):
        print(f"{self.নাম} ঘেউ ঘেউ করছে")

class বিড়াল(প্রাণী):
    def ডাকে(self):
        print(f"{self.নাম} মিউ মিউ করছে")

tommy = কুকুর("Tommy")
tommy.শ্বাস_নেয়()  # প্রাণী থেকে পেয়েছে
tommy.ডাকে()        # নিজের

kitty = বিড়াল("Kitty")
kitty.শ্বাস_নেয়()  # প্রাণী থেকে পেয়েছে
kitty.ডাকে()        # নিজের
```

> 👨‍👩‍👦 উদাহরণ: **বাবার সম্পত্তি ছেলে পায় — তেমনি Parent Class-এর সব কিছু Child Class পায়**

---

### 🏛️ 4. Polymorphism (বহুরূপিতা)

**একই নামের কাজ, কিন্তু আলাদা আলাদা ভাবে হয়।**

```python
class বিড়াল:
    def শব্দ_করো(self):
        print("মিউ মিউ 🐱")

class কুকুর:
    def শব্দ_করো(self):
        print("ঘেউ ঘেউ 🐕")

class গরু:
    def শব্দ_করো(self):
        print("হাম্বা 🐄")

# একই function — আলাদা কাজ
প্রাণীর_তালিকা = [বিড়াল(), কুকুর(), গরু()]

for প্রাণী in প্রাণীর_তালিকা:
    প্রাণী.শব্দ_করো()

# Output:
# মিউ মিউ 🐱
# ঘেউ ঘেউ 🐕
# হাম্বা 🐄
```

> 🎭 উদাহরণ: **"আঁকো" বললে — কেউ ছবি আঁকে, কেউ ডিজাইন করে, কেউ sketch করে — কাজ আলাদা কিন্তু command একই**

# 3️⃣ Functional Programming (কার্যকরী প্রোগ্রামিং)

## 🔵 সহজ ভাষায়:
**সব কিছু Function দিয়ে করো, কোনো Data পরিবর্তন করো না।**

### মূল নিয়ম:
- **Pure Function** → একই input দিলে সবসময় একই output
- **Immutability** → Data একবার তৈরি হলে পরিবর্তন করা যাবে না
- **No Side Effects** → Function শুধু তার কাজ করবে, বাইরে কিছু প্রভাবিত করবে না

## 👨‍💻 Code উদাহরণ:

```python
# ❌ Impure Function (খারাপ)
total = 0
def যোগ_করো(সংখ্যা):
    global total       # বাইরের variable পরিবর্তন করছে — এটা ভালো না
    total += সংখ্যা
    return total

# ✅ Pure Function (ভালো)
def যোগ_করো(ক, খ):
    return ক + খ      # শুধু নিজের কাজ, বাইরে কিছু স্পর্শ করে না

print(যোগ_করো(3, 4))  # সবসময় 7
print(যোগ_করো(3, 4))  # সবসময় 7 ✅
```

### Functional Programming-এর শক্তিশালী tools:

```python
সংখ্যা = [1, 2, 3, 4, 5]

# Map → প্রতিটা element-এ কাজ করো
দ্বিগুণ = list(map(lambda x: x * 2, সংখ্যা))
print(দ্বিগুণ)  # [2, 4, 6, 8, 10]

# Filter → শর্ত অনুযায়ী বাছাই করো
জোড় = list(filter(lambda x: x % 2 == 0, সংখ্যা))
print(জোড়)  # [2, 4]

# Reduce → সব মিলিয়ে একটা ফলাফল
from functools import reduce
গুণফল = reduce(lambda x, y: x * y, সংখ্যা)
print(গুণফল)  # 120
```

# 🍕 বাস্তব উদাহরণ দিয়ে বুঝি — Pizza বানানো

```python
# ✅ Procedural Style
আটা = 500
পনির = 200
সস = 100

মিশাও(আটা, পনির, সস)
বেক_করো(30)  # মিনিট
print("Pizza তৈরি!")

# ✅ OOP Style
class Pizza:
    def __init__(self, আটা, পনির):
        self.আটা = আটা
        self.পনির = পনির
    
    def বানাও(self):
        print(f"{self.আটা}g আটা ও {self.পনির}g পনির দিয়ে Pizza তৈরি!")

আমার_pizza = Pizza(500, 200)
আমার_pizza.বানাও()

# ✅ Functional Style
def pizza_বানাও(আটা, পনির, সস):
    return f"Pizza: {আটা}g আটা, {পনির}g পনির, {সস}g সস"

ফলাফল = pizza_বানাও(500, 200, 100)
print(ফলাফল)
```

## 2️⃣ Interpreted Language (ইন্টারপ্রেটেড ভাষা)

### 🔍 Interpreted মানে কী?

- **Compiled Language** → আগে পুরো কোড **compile** করতে হয়, তারপর run হয় *(C, C++)*
- **Interpreted Language** → কোড **line by line** সরাসরি run হয় *(Python)*

```
Compiled:   কোড লেখো → Compile করো → .exe তৈরি → Run করো
Interpreted: কোড লেখো → সরাসরি Run করো ✅ (সহজ!)
```

```python
print("লাইন ১")   # আগে এটা চলে
print("লাইন ২")   # তারপর এটা
print("লাইন ৩")   # তারপর এটা
# একটায় error থাকলে সেখানে গিয়ে থামে
```

> ✅ সুবিধা: সাথে সাথে result দেখা যায়, debug করা সহজ
> ❌ অসুবিধা: Compiled language-এর চেয়ে একটু ধীর

## 1️⃣ Indentation (ইন্ডেন্টেশন)

```python
# ❌ ভুল — 2 space
def hello():
  print("Hello")

# ❌ ভুল — Tab ব্যবহার
def hello():
	print("Hello")

# ✅ সঠিক — ৪টি space
def hello():
    print("Hello")  # ৪ space ✅
    if True:
        print("Nested")  # আরও ৪ space = মোট ৮ ✅
```

# 📋 PEP 8 — সম্পূর্ণ সারসংক্ষেপ

```python
# ✅ PEP 8 মেনে লেখা Perfect Code

import os
import sys

import numpy as np  # Third-party

MAX_VALUE = 100  # Constant


class BankAccount:
    """ব্যাংক একাউন্ট ম্যানেজ করার class।"""

    def __init__(self, owner_name, balance=0):
        self.owner_name = owner_name
        self._balance = balance  # Private

    def deposit(self, amount):
        """টাকা জমা দেওয়ার method।"""
        if amount > 0:
            self._balance += amount
            return True
        return False

    def get_balance(self):
        """বর্তমান balance দেখানোর method।"""
        return self._balance


def calculate_interest(principal, rate, time):
    """সুদ হিসাব করার function।"""
    interest = principal * rate * time / 100
    return interest


# Main code
account = BankAccount("Rahim", 5000)
account.deposit(2000)

interest = calculate_interest(7000, 5, 2)
print(f"সুদ: {interest}")
```

## 🔷 int (Integer — পূর্ণ সংখ্যা)

দশমিক ছাড়া যেকোনো সংখ্যা।

```python
বয়স = 25
তাপমাত্রা = -10      # ঋণাত্মক সংখ্যাও হয়
জনসংখ্যা = 170000000  # অনেক বড় সংখ্যাও হয়

print(type(বয়স))  # <class 'int'>

# গাণিতিক কাজ
print(10 + 3)   # 13  → যোগ
print(10 - 3)   # 7   → বিয়োগ
print(10 * 3)   # 30  → গুণ
print(10 // 3)  # 3   → ভাগ (পূর্ণ সংখ্যা)
print(10 % 3)   # 1   → ভাগশেষ
print(10 ** 3)  # 1000 → ঘাত (10 এর ঘনফল)

# 💡 Python-এ int-এর কোনো সীমা নেই!
বিশাল_সংখ্যা = 99999999999999999999999999
print(বিশাল_সংখ্যা)  # কোনো সমস্যা নেই ✅
```

# 4️⃣ List (তালিকা) — সবচেয়ে বেশি ব্যবহৃত!

```python
ফলের_তালিকা = ["আম", "কলা", "লিচু"]
মিশ্র = [1, "Hello", 3.14, True]  # যেকোনো type রাখা যায়!
খালি = []

print(type(ফলের_তালিকা))  # <class 'list'>
```

## 📋 List-এর গুরুত্বপূর্ণ কাজ:

```python
সংখ্যা = [10, 20, 30, 40, 50]

# Access — নির্দিষ্ট item নেওয়া
print(সংখ্যা[0])    # 10 (প্রথম)
print(সংখ্যা[-1])   # 50 (শেষ)
print(সংখ্যা[1:3])  # [20, 30] (Slicing)

# পরিবর্তন করা যায়! (Mutable)
সংখ্যা[0] = 100
print(সংখ্যা)  # [100, 20, 30, 40, 50]

# যোগ করা
সংখ্যা.append(60)      # শেষে যোগ
সংখ্যা.insert(0, 5)    # নির্দিষ্ট জায়গায় যোগ

# বাদ দেওয়া
সংখ্যা.remove(20)      # মান দিয়ে বাদ
সংখ্যা.pop()           # শেষেরটা বাদ
সংখ্যা.pop(0)          # নির্দিষ্ট index বাদ

# খোঁজা
print(30 in সংখ্যা)    # True/False
print(সংখ্যা.index(30)) # কোন position-এ আছে

# সাজানো
সংখ্যা.sort()          # ছোট থেকে বড়
সংখ্যা.sort(reverse=True)  # বড় থেকে ছোট

# দৈর্ঘ্য
print(len(সংখ্যা))     # কতটা item আছে

# Loop
for item in সংখ্যা:
    print(item)
```

# 5️⃣ Tuple (অপরিবর্তনযোগ্য তালিকা)

**List-এর মতো, কিন্তু একবার তৈরি হলে পরিবর্তন করা যায় না।**

```python
রঙ = ("লাল", "সবুজ", "নীল")
স্থানাঙ্ক = (23.8, 90.4)  # GPS coordinate

print(type(রঙ))  # <class 'tuple'>

# Access করা যায়
print(রঙ[0])   # লাল
print(রঙ[-1])  # নীল

# ❌ পরিবর্তন করা যায় না!
# রঙ[0] = "হলুদ"  → Error!

# Tuple unpacking — দারুণ feature!
x, y = (10, 20)
print(x)  # 10
print(y)  # 20

# Function থেকে একাধিক value return
def min_max(numbers):
    return min(numbers), max(numbers)

ন্যূনতম, সর্বোচ্চ = min_max([3, 1, 4, 1, 5, 9])
print(ন্যূনতম)  # 1
print(সর্বোচ্চ)  # 9
```

```python
ছাত্র = {
    "নাম": "রহিম",
    "বয়স": 20,
    "শহর": "ঢাকা",
    "CGPA": 3.85
}

print(type(ছাত্র))  # <class 'dict'>
```

## 📖 Dictionary-এর গুরুত্বপূর্ণ কাজ:

```python
ব্যক্তি = {"নাম": "করিম", "বয়স": 25, "শহর": "চট্টগ্রাম"}

# Value নেওয়া
print(ব্যক্তি["নাম"])          # করিম
print(ব্যক্তি.get("বয়স"))     # 25
print(ব্যক্তি.get("ফোন", "নেই"))  # নেই (না থাকলে default)

# যোগ / পরিবর্তন
ব্যক্তি["ফোন"] = "01712345678"  # নতুন key যোগ
ব্যক্তি["বয়স"] = 26            # পরিবর্তন

# বাদ দেওয়া
del ব্যক্তি["শহর"]
ব্যক্তি.pop("ফোন")

# সব key দেখা
print(ব্যক্তি.keys())    # dict_keys(['নাম', 'বয়স'])

# সব value দেখা
print(ব্যক্তি.values())  # dict_values(['করিম', 26])

# সব জোড়া দেখা
print(ব্যক্তি.items())   # dict_items([('নাম', 'করিম'), ...])

# Loop
for key, value in ব্যক্তি.items():
    print(f"{key}: {value}")

# Check করা
print("নাম" in ব্যক্তি)  # True
```

# 7️⃣ Set (অনন্য সংগ্রহ)

**Duplicate নেই, order নেই — শুধু unique value।**

```python
ফল = {"আম", "কলা", "লিচু", "আম", "কলা"}
print(ফল)  # {'আম', 'কলা', 'লিচু'} — duplicate বাদ! ✅

print(type(ফল))  # <class 'set'>
```

## 🔵 Set-এর গুরুত্বপূর্ণ কাজ:

```python
ক্লাস_এ = {"রহিম", "করিম", "জলিল"}
ক্লাস_বি = {"করিম", "সালাম", "জলিল"}

# যোগ / বাদ
ক্লাস_এ.add("নতুন")
ক্লাস_এ.remove("রহিম")

# Set Operations — গণিতের মতো!
# Union (∪) — দুটো মিলিয়ে
print(ক্লাস_এ | ক্লাস_বি)  # সব ছাত্র

# Intersection (∩) — উভয়তে আছে
print(ক্লাস_এ & ক্লাস_বি)  # শুধু করিম, জলিল

# Difference (-) — শুধু এতে আছে
print(ক্লাস_এ - ক্লাস_বি)  # শুধু ক্লাস এ-তে আছে

# Symmetric Difference — যেকোনো একটায় আছে কিন্তু উভয়তে নেই
print(ক্লাস_এ ^ ক্লাস_বি)

# 💡 Duplicate বাদ দেওয়ার সহজ উপায়!
তালিকা = [1, 2, 2, 3, 3, 3, 4]
unique = list(set(তালিকা))
print(unique)  # [1, 2, 3, 4] ✅
```

# 9️⃣ None Type

**কোনো মান নেই — এটাই বোঝায়।**

```python
ফলাফল = None

print(type(ফলাফল))   # <class 'NoneType'>
print(ফলাফল is None) # True

# কখন ব্যবহার হয়?
def কিছু_খোঁজো(তালিকা, লক্ষ্য):
    for item in তালিকা:
        if item == লক্ষ্য:
            return item
    return None  # না পেলে None return করো

ফল = কিছু_খোঁজো([1, 2, 3], 5)
if ফল is None:
    print("পাওয়া যায়নি")
```

# 🎯 Interview-এ যা জিজ্ঞেস করতে পারে:

**Q: Mutable vs Immutable কী?**
> ✅ **Mutable** = তৈরির পরে পরিবর্তন করা যায় → `list, dict, set`
> ✅ **Immutable** = পরিবর্তন করা যায় না → `int, float, str, tuple`

**Q: List vs Tuple পার্থক্য কী?**
> ✅ List পরিবর্তন করা যায়, Tuple যায় না। Tuple দ্রুত ও কম memory নেয়।

**Q: Dictionary-তে কি Duplicate key হয়?**
> ✅ না! Dictionary-তে key সবসময় unique। একই key দুবার দিলে পরেরটা আগেরটাকে replace করে।

**Q: Set-এ কেন order নেই?**
> ✅ Set internally **hash table** ব্যবহার করে, তাই insertion order বজায় রাখে না।

**Q: `None` এবং `False` কি একই?**
> ✅ না। `None` মানে কোনো মান নেই, `False` মানে মিথ্যা। তবে উভয়ই **Falsy** — if condition-এ False হিসেবে কাজ করে।

**Q: String কেন Immutable?**
> ✅ Security ও performance-এর জন্য। String immutable হওয়ায় Python একই string বারবার তৈরি না করে **reuse** করতে পারে।

# 1️⃣ সবচেয়ে বড় পার্থক্য — Mutable vs Immutable

## ✅ List — Mutable (পরিবর্তনযোগ্য)

```python
ফল = ["আম", "কলা", "লিচু"]

# পরিবর্তন করা যায় ✅
ফল[0] = "আনারস"
print(ফল)  # ['আনারস', 'কলা', 'লিচু']

# যোগ করা যায় ✅
ফল.append("আঙুর")
print(ফল)  # ['আনারস', 'কলা', 'লিচু', 'আঙুর']

# বাদ দেওয়া যায় ✅
ফল.remove("কলা")
print(ফল)  # ['আনারস', 'লিচু', 'আঙুর']
```

## ❌ Tuple — Immutable (অপরিবর্তনযোগ্য)

```python
রঙ = ("লাল", "সবুজ", "নীল")

# পরিবর্তন করা যায় না ❌
# রঙ[0] = "হলুদ"  → TypeError: 'tuple' object does not support item assignment

# যোগ করা যায় না ❌
# রঙ.append("বেগুনি")  → AttributeError: 'tuple' has no attribute 'append'

# বাদ দেওয়া যায় না ❌
# রঙ.remove("লাল")  → AttributeError

# তবে পুরো tuple replace করা যায়
রঙ = ("হলুদ", "গোলাপী")  # এটা নতুন tuple তৈরি করে
print(রঙ)  # ('হলুদ', 'গোলাপী')
```

# 2️⃣ Speed (গতি) — Tuple দ্রুত!

```python
import timeit

# List কত সময় নেয়?
list_time = timeit.timeit(
    stmt="[1, 2, 3, 4, 5]",
    number=1000000
)

# Tuple কত সময় নেয়?
tuple_time = timeit.timeit(
    stmt="(1, 2, 3, 4, 5)",
    number=1000000
)

print(f"List সময়:  {list_time:.4f} সেকেন্ড")
print(f"Tuple সময়: {tuple_time:.4f} সেকেন্ড")

# Output (প্রায়):
# List সময়:  0.0589 সেকেন্ড
# Tuple সময়: 0.0187 সেকেন্ড ← প্রায় ৩ গুণ দ্রুত! ✅
```

> 💡 **কেন Tuple দ্রুত?**
> Tuple immutable, তাই Python জানে এটা বদলাবে না।
> তাই memory-তে একটা নির্দিষ্ট জায়গায় রাখে — access করা সহজ হয়!

# 6️⃣ Tuple Unpacking — দারুণ Feature!

```python
# Tuple unpacking
নাম, বয়স, শহর = ("রহিম", 25, "ঢাকা")
print(নাম)   # রহিম
print(বয়স)  # 25
print(শহর)  # ঢাকা

# Swap করা — Tuple দিয়ে সহজে!
a = 10
b = 20
a, b = b, a  # আসলে (b, a) একটা tuple!
print(a, b)  # 20 10

# Function থেকে একাধিক value return
def বিভাজন(সংখ্যা, ভাজক):
    ভাগফল = সংখ্যা // ভাজক
    ভাগশেষ = সংখ্যা % ভাজক
    return ভাগফল, ভাগশেষ  # Tuple return করছে!

ফল, বাকি = বিভাজন(17, 5)
print(f"ভাগফল: {ফল}, ভাগশেষ: {বাকি}")
# ভাগফল: 3, ভাগশেষ: 2

# * দিয়ে বাকিগুলো নেওয়া
প্রথম, *মাঝের, শেষ = (1, 2, 3, 4, 5)
print(প্রথম)   # 1
print(মাঝের)  # [2, 3, 4]
print(শেষ)    # 5
```

# 7️⃣ Nested (ভেতরে ভেতরে)

```python
# Nested List — পরিবর্তন করা যায়
ম্যাট্রিক্স = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
ম্যাট্রিক্স[0][0] = 100  # পরিবর্তন ✅
print(ম্যাট্রিক্স[0])  # [100, 2, 3]

# Nested Tuple
বিন্দু = ((1, 2), (3, 4), (5, 6))
print(বিন্দু[0])     # (1, 2)
print(বিন্দু[0][1])  # 2

# ⚠️ মজার বিষয়!
# Tuple-এর ভেতরে List থাকলে সেই List পরিবর্তন করা যায়!
মিশ্র = ([1, 2, 3], "Hello", 42)
মিশ্র[0].append(4)  # ✅ List পরিবর্তন হয়েছে!
print(মিশ্র)  # ([1, 2, 3, 4], 'Hello', 42)
# মিশ্র[1] = "World"  ❌ Tuple নিজে পরিবর্তন হয় না
```

# 8️⃣ কখন কোনটা ব্যবহার করবো?

# 🎯 Interview-এ যা জিজ্ঞেস করতে পারে:

**Q: List ও Tuple-এর মূল পার্থক্য কী?**
> ✅ List **mutable** (পরিবর্তন যোগ্য), Tuple **immutable** (অপরিবর্তনযোগ্য)। List-এ `[ ]` আর Tuple-এ `( )` ব্যবহার হয়।

**Q: কখন Tuple ব্যবহার করবো?**
> ✅ যখন data **স্থায়ী** — যেমন GPS coordinate, জন্মতারিখ, RGB color। এছাড়া Dictionary-তে key হিসেবে এবং Function থেকে একাধিক value return করতে।

**Q: Tuple কি সত্যিই দ্রুত?**
> ✅ হ্যাঁ। Tuple immutable বলে Python সেটা **optimize** করতে পারে। তাই access এবং তৈরি করা List-এর চেয়ে প্রায় ৩ গুণ দ্রুত।

**Q: Tuple-এর ভেতরে List থাকলে কি পরিবর্তন করা যায়?**
> ✅ হ্যাঁ! Tuple নিজে পরিবর্তন হয় না, কিন্তু ভেতরের List পরিবর্তন করা যায়। কারণ Tuple শুধু **reference** ধরে রাখে, List-এর content নয়।

**Q: একটা item-এর Tuple কীভাবে বানাবো?**
> ✅ `(1,)` — শেষে কমা লাগবে। `(1)` লিখলে সেটা tuple না, শুধু integer।

**Q: Tuple unpacking কী?**
> ✅ Tuple-এর একাধিক মান একসাথে আলাদা variable-এ রাখা। যেমন: `a, b, c = (1, 2, 3)`

**Dictionary হলো Key-Value জোড়ায় data রাখার একটি পাত্র।**

> 🗣️ সহজ উদাহরণ:

> ঠিক তেমনি Python-এর Dictionary-তে —

# 🏗️ Dictionary তৈরির পদ্ধতি

```python
# পদ্ধতি ১ — সরাসরি { } দিয়ে
ব্যক্তি = {"নাম": "করিম", "বয়স": 25}

# পদ্ধতি ২ — dict() function দিয়ে
ব্যক্তি = dict(নাম="করিম", বয়স=25)

# পদ্ধতি ৩ — খালি dictionary তৈরি করে পরে যোগ
ব্যক্তি = {}
ব্যক্তি["নাম"] = "করিম"
ব্যক্তি["বয়স"] = 25

# পদ্ধতি ৪ — dict() + list of tuples
ব্যক্তি = dict([("নাম", "করিম"), ("বয়স", 25)])

# পদ্ধতি ৫ — fromkeys() দিয়ে
# সব key-এর জন্য একই default value
চাবি = ["নাম", "বয়স", "শহর"]
খালি_তথ্য = dict.fromkeys(চাবি, "অজানা")
print(খালি_তথ্য)
# {'নাম': 'অজানা', 'বয়স': 'অজানা', 'শহর': 'অজানা'}

print(type(ব্যক্তি))  # <class 'dict'>
```

# 1️⃣ Dictionary থেকে Value নেওয়া (Access)

```python
ছাত্র = {
    "নাম": "রহিম",
    "বয়স": 20,
    "শহর": "ঢাকা"
}

# পদ্ধতি ১ — [ ] দিয়ে
print(ছাত্র["নাম"])   # রহিম
print(ছাত্র["বয়স"])  # 20

# ⚠️ সমস্যা — key না থাকলে Error!
# print(ছাত্র["ফোন"])  # KeyError: 'ফোন' ❌

# পদ্ধতি ২ — .get() দিয়ে (নিরাপদ উপায়!)
print(ছাত্র.get("নাম"))          # রহিম
print(ছাত্র.get("ফোন"))          # None (Error নেই ✅)
print(ছাত্র.get("ফোন", "নেই"))   # নেই (Default value ✅)

# 💡 সবসময় .get() ব্যবহার করা ভালো অভ্যাস!
```

# 2️⃣ Dictionary-তে যোগ ও পরিবর্তন

```python
ব্যক্তি = {"নাম": "রহিম", "বয়স": 20}

# নতুন key-value যোগ
ব্যক্তি["শহর"] = "ঢাকা"
ব্যক্তি["ফোন"] = "01711111111"
print(ব্যক্তি)
# {'নাম': 'রহিম', 'বয়স': 20, 'শহর': 'ঢাকা', 'ফোন': '01711111111'}

# পরিবর্তন করা
ব্যক্তি["বয়স"] = 21       # বয়স আপডেট
ব্যক্তি["নাম"] = "করিম"   # নাম আপডেট
print(ব্যক্তি["বয়স"])  # 21

# .update() দিয়ে একসাথে অনেক কিছু যোগ/পরিবর্তন
ব্যক্তি.update({
    "বয়স": 22,
    "পেশা": "Developer",
    "দেশ": "বাংলাদেশ"
})
print(ব্যক্তি)
```

# 3️⃣ Dictionary থেকে বাদ দেওয়া (Delete)

```python
তথ্য = {
    "নাম": "রহিম",
    "বয়স": 20,
    "শহর": "ঢাকা",
    "ফোন": "01711111111"
}

# পদ্ধতি ১ — del দিয়ে
del তথ্য["ফোন"]
print(তথ্য)  # ফোন বাদ গেছে

# পদ্ধতি ২ — .pop() দিয়ে (value ফেরত দেয়!)
বয়স = তথ্য.pop("বয়স")
print(বয়স)   # 20 — বাদ দেওয়া value পেলাম
print(তথ্য)  # বয়স বাদ গেছে

# .pop() — key না থাকলে default return করে
ফোন = তথ্য.pop("ফোন", "নেই")
print(ফোন)  # নেই (Error নেই ✅)

# পদ্ধতি ৩ — .popitem() — শেষের item বাদ দেয়
শেষ_item = তথ্য.popitem()
print(শেষ_item)  # ('শহর', 'ঢাকা') — tuple হিসেবে পায়

# পদ্ধতি ৪ — .clear() — সব বাদ দাও
তথ্য.clear()
print(তথ্য)  # {} — খালি!
```

# 4️⃣ Dictionary Loop করা

```python
ছাত্র = {
    "নাম": "রহিম",
    "বয়স": 20,
    "শহর": "ঢাকা",
    "CGPA": 3.85
}

# শুধু Key দেখা
print("--- Keys ---")
for key in ছাত্র:
    print(key)
# নাম, বয়স, শহর, CGPA

# শুধু Value দেখা
print("--- Values ---")
for value in ছাত্র.values():
    print(value)
# রহিম, 20, ঢাকা, 3.85

# Key ও Value একসাথে দেখা
print("--- Key & Value ---")
for key, value in ছাত্র.items():
    print(f"{key}: {value}")
# নাম: রহিম
# বয়স: 20
# শহর: ঢাকা
# CGPA: 3.85
```

# 5️⃣ গুরুত্বপূর্ণ Methods

```python
তথ্য = {"নাম": "রহিম", "বয়স": 20, "শহর": "ঢাকা"}

# .keys() — সব key দেখো
print(তথ্য.keys())
# dict_keys(['নাম', 'বয়স', 'শহর'])

# .values() — সব value দেখো
print(তথ্য.values())
# dict_values(['রহিম', 20, 'ঢাকা'])

# .items() — সব key-value জোড়া দেখো
print(তথ্য.items())
# dict_items([('নাম', 'রহিম'), ('বয়স', 20), ('শহর', 'ঢাকা')])

# .copy() — copy তৈরি করো
নতুন_তথ্য = তথ্য.copy()

# len() — কতটা item আছে
print(len(তথ্য))  # 3

# in — key আছে কিনা চেক করো
print("নাম" in তথ্য)    # True ✅
print("ফোন" in তথ্য)   # False ❌
print("রহিম" in তথ্য)  # False ❌ (value চেক হয় না!)

# value চেক করতে:
print("রহিম" in তথ্য.values())  # True ✅
```

# 6️⃣ Nested Dictionary (ভেতরে ভেতরে Dictionary)

```python
# Dictionary-র ভেতরে Dictionary!
স্কুল = {
    "ছাত্র_১": {
        "নাম": "রহিম",
        "বয়স": 15,
        "নম্বর": {
            "বাংলা": 85,
            "ইংরেজি": 90,
            "গণিত": 95
        }
    },
    "ছাত্র_২": {
        "নাম": "করিম",
        "বয়স": 16,
        "নম্বর": {
            "বাংলা": 78,
            "ইংরেজি": 82,
            "গণিত": 88
        }
    }
}

# Nested access
print(স্কুল["ছাত্র_১"]["নাম"])                # রহিম
print(স্কুল["ছাত্র_১"]["নম্বর"]["গণিত"])       # 95
print(স্কুল["ছাত্র_২"]["নম্বর"]["ইংরেজি"])     # 82

# সব ছাত্রের তথ্য দেখানো
for ছাত্র_id, তথ্য in স্কুল.items():
    print(f"\n{ছাত্র_id}:")
    print(f"  নাম: {তথ্য['নাম']}")
    print(f"  গণিত: {তথ্য['নম্বর']['গণিত']}")
```

# 7️⃣ Dictionary Comprehension

```python
# সাধারণ উপায়
বর্গ = {}
for i in range(1, 6):
    বর্গ[i] = i ** 2
print(বর্গ)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Dictionary Comprehension — মাত্র ১ লাইন! ✅
বর্গ = {i: i ** 2 for i in range(1, 6)}
print(বর্গ)  # {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

# Condition সহ
জোড়_বর্গ = {i: i ** 2 for i in range(1, 11) if i % 2 == 0}
print(জোড়_বর্গ)  # {2: 4, 4: 16, 6: 36, 8: 64, 10: 100}

# দুটো List থেকে Dictionary
নাম = ["রহিম", "করিম", "জলিল"]
নম্বর = ["01711", "01722", "01733"]

ফোনবুক = {নাম[i]: নম্বর[i] for i in range(len(নাম))}
# অথবা আরও সহজে:
ফোনবুক = dict(zip(নাম, নম্বর))
print(ফোনবুক)
# {'রহিম': '01711', 'করিম': '01722', 'জলিল': '01733'}
```

# 9️⃣ বাস্তব জীবনের উদাহরণ

```python
# ✅ উদাহরণ ১ — Student Result System
def ফলাফল_দেখাও(ছাত্র):
    মোট = sum(ছাত্র["নম্বর"].values())
    গড় = মোট / len(ছাত্র["নম্বর"])
    print(f"নাম: {ছাত্র['নাম']}")
    print(f"মোট নম্বর: {মোট}")
    print(f"গড়: {গড়:.2f}")
    print(f"গ্রেড: {'A+' if গড় >= 90 else 'A' if গড় >= 80 else 'B'}")

রহিম = {
    "নাম": "রহিম",
    "নম্বর": {"বাংলা": 85, "ইংরেজি": 90, "গণিত": 95}
}
ফলাফল_দেখাও(রহিম)


# ✅ উদাহরণ ২ — Word Count করা
বাক্য = "আম কলা আম লিচু কলা আম"
শব্দ_গণনা = {}

for শব্দ in বাক্য.split():
    শব্দ_গণনা[শব্দ] = শব্দ_গণনা.get(শব্দ, 0) + 1

print(শব্দ_গণনা)
# {'আম': 3, 'কলা': 2, 'লিচু': 1}


# ✅ উদাহরণ ৩ — Config/Settings
app_config = {
    "app_name": "MyApp",
    "version": "1.0.0",
    "database": {
        "host": "localhost",
        "port": 5432,
        "name": "mydb"
    },
    "debug": True,
    "max_connections": 100
}

print(app_config["database"]["host"])  # localhost
print(app_config["version"])           # 1.0.0
```

# 📊 Dictionary-র সব Method একনজরে

| Method | কাজ | উদাহরণ |
|--------|-----|---------|
| `.get(key)` | নিরাপদে value নাও | `d.get("নাম", "নেই")` |
| `.keys()` | সব key দেখো | `d.keys()` |
| `.values()` | সব value দেখো | `d.values()` |
| `.items()` | সব জোড়া দেখো | `d.items()` |
| `.update()` | যোগ/পরিবর্তন | `d.update({"বয়স": 25})` |
| `.pop(key)` | বাদ দাও, value পাও | `d.pop("নাম")` |
| `.popitem()` | শেষ item বাদ দাও | `d.popitem()` |
| `.clear()` | সব বাদ দাও | `d.clear()` |
| `.copy()` | copy তৈরি | `d.copy()` |

# 🎯 Interview-এ যা জিজ্ঞেস করতে পারে:

**Q: Dictionary কী এবং কীভাবে কাজ করে?**
> ✅ Dictionary হলো Key-Value জোড়ায় data রাখার structure। Key দিয়ে খুব দ্রুত Value খোঁজা যায়। এটা internally **Hash Table** ব্যবহার করে।

**Q: Dictionary-তে `[ ]` ও `.get()` এর পার্থক্য কী?**
> ✅ `[ ]` দিয়ে key না থাকলে **KeyError** আসে। `.get()` দিয়ে key না থাকলে **None** বা **default value** পাওয়া যায় — কোনো Error নেই।

**Q: Dictionary কি Ordered?**
> ✅ **Python 3.7+** থেকে Dictionary **insertion order** বজায় রাখে। তার আগে unordered ছিল।

**Q: Dictionary-র Key হিসেবে কী ব্যবহার করা যায়?**
> ✅ যেকোনো **Immutable** type — String, Integer, Float, Tuple। List বা Dictionary key হতে পারে না কারণ এগুলো **mutable** তাই **unhashable**।

**Q: Dictionary internally কীভাবে কাজ করে?**
> ✅ Dictionary **Hash Table** ব্যবহার করে। Key-কে **hash()** function দিয়ে একটা সংখ্যায় রূপান্তর করে এবং সেই অনুযায়ী memory-তে রাখে। এ কারণে Dictionary-তে lookup **O(1)** — অর্থাৎ যত বড়ই হোক, খোঁজা সমান দ্রুত!

**Q: Dictionary ও List-এর পার্থক্য কী?**
> ✅ List-এ **index** (0, 1, 2...) দিয়ে access করা হয়। Dictionary-তে **key** দিয়ে access করা হয়। Dictionary data খুঁজতে দ্রুত কিন্তু List-এর চেয়ে বেশি memory নেয়।

## পদ্ধতি ১ — Curly Bracket `{ }` দিয়ে

```python
# সরাসরি { } দিয়ে তৈরি
ফল = {"আম", "কলা", "লিচু"}
সংখ্যা = {1, 2, 3, 4, 5}
মিশ্র = {1, "Hello", 3.14, True}

print(ফল)           # {'আম', 'কলা', 'লিচু'}
print(type(ফল))     # <class 'set'>

# ⚠️ Duplicate দিলে স্বয়ংক্রিয়ভাবে বাদ যায়!
সংখ্যা = {1, 2, 2, 3, 3, 3, 4}
print(সংখ্যা)  # {1, 2, 3, 4} — duplicate বাদ ✅
```

## পদ্ধতি ২ — `set()` Function দিয়ে

```python
# List থেকে Set
তালিকা = [1, 2, 2, 3, 3, 4]
আমার_set = set(তালিকা)
print(আমার_set)  # {1, 2, 3, 4}

# String থেকে Set — প্রতিটা অক্ষর আলাদা হয়!
অক্ষর = set("Bangladesh")
print(অক্ষর)
# {'B', 'a', 'n', 'g', 'l', 'd', 'e', 's', 'h'}
# duplicate 'a' একবারই আছে!

# Tuple থেকে Set
রঙ_tuple = ("লাল", "সবুজ", "লাল", "নীল")
রঙ_set = set(রঙ_tuple)
print(রঙ_set)  # {'লাল', 'সবুজ', 'নীল'}

# Range থেকে Set
পাঁচ = set(range(1, 6))
print(পাঁচ)  # {1, 2, 3, 4, 5}
```

## পদ্ধতি ৪ — Set Comprehension

```python
# List Comprehension-এর মতোই কাজ করে
বর্গ = {x ** 2 for x in range(1, 6)}
print(বর্গ)  # {1, 4, 9, 16, 25}

# Condition সহ
জোড় = {x for x in range(1, 11) if x % 2 == 0}
print(জোড়)  # {2, 4, 6, 8, 10}

# String থেকে unique vowel
বাক্য = "Hello Bangladesh"
vowels = {ch.lower() for ch in বাক্য if ch.lower() in "aeiou"}
print(vowels)  # {'e', 'a', 'i'}
```

# 1️⃣ Set-এ যোগ ও বাদ দেওয়া

```python
ফল = {"আম", "কলা", "লিচু"}

# একটা item যোগ — .add()
ফল.add("আঙুর")
print(ফল)  # {'আম', 'কলা', 'লিচু', 'আঙুর'}

# একই item আবার যোগ করলে কিছু হয় না!
ফল.add("আম")
print(ফল)  # {'আম', 'কলা', 'লিচু', 'আঙুর'} — একই ✅

# একাধিক item যোগ — .update()
ফল.update(["আনারস", "পেয়ারা", "কলা"])
print(ফল)
# duplicate 'কলা' যোগ হয়নি!

# একটা item বাদ — .remove()
ফল.remove("কলা")
print(ফল)  # কলা বাদ গেছে
# ⚠️ item না থাকলে KeyError আসে!

# নিরাপদে বাদ দাও — .discard()
ফল.discard("কলা")   # কলা নেই, তবুও Error নেই ✅
ফল.discard("আম")    # আম বাদ গেল

# যেকোনো একটা বাদ দাও — .pop()
বাদ_যাওয়া = ফল.pop()  # কোনটা বাদ যাবে অনিশ্চিত!
print(বাদ_যাওয়া)

# সব বাদ দাও — .clear()
ফল.clear()
print(ফল)  # set()
```

# 2️⃣ Set Operations — গণিতের মতো! ⭐

এটাই Set-এর **সবচেয়ে শক্তিশালী** দিক!

```python
ক্লাস_এ = {"রহিম", "করিম", "জলিল", "সালাম"}
ক্লাস_বি = {"করিম", "সালাম", "মামুন", "রফিক"}
```

## 🔶 Union (∪) — দুটো মিলিয়ে সব

```python
# পদ্ধতি ১ — | operator
সব = ক্লাস_এ | ক্লাস_বি

# পদ্ধতি ২ — .union()
সব = ক্লাস_এ.union(ক্লাস_বি)

print(সব)
# {'রহিম', 'করিম', 'জলিল', 'সালাম', 'মামুন', 'রফিক'}
# duplicate নেই! ✅
```

## 🔷 Intersection (∩) — উভয়তেই আছে

```python
# পদ্ধতি ১ — & operator
উভয়তে = ক্লাস_এ & ক্লাস_বি

# পদ্ধতি ২ — .intersection()
উভয়তে = ক্লাস_এ.intersection(ক্লাস_বি)

print(উভয়তে)
# {'করিম', 'সালাম'} — শুধু এরা দুই ক্লাসেই আছে
```

## 🔴 Difference (-) — একটায় আছে, অন্যটায় নেই

```python
# পদ্ধতি ১ — - operator
শুধু_এ = ক্লাস_এ - ক্লাস_বি

# পদ্ধতি ২ — .difference()
শুধু_এ = ক্লাস_এ.difference(ক্লাস_বি)

print(শুধু_এ)
# {'রহিম', 'জলিল'} — শুধু ক্লাস এ-তে আছে
print(ক্লাস_বি - ক্লাস_এ)
# {'মামুন', 'রফিক'} — শুধু ক্লাস বি-তে আছে
```

## 🟡 Symmetric Difference (△) — যেকোনো একটায় আছে, কিন্তু উভয়তে নেই

```python
# পদ্ধতি ১ — ^ operator
যেকোনো_একটায় = ক্লাস_এ ^ ক্লাস_বি

# পদ্ধতি ২ — .symmetric_difference()
যেকোনো_একটায় = ক্লাস_এ.symmetric_difference(ক্লাস_বি)

print(যেকোনো_একটায়)
# {'রহিম', 'জলিল', 'মামুন', 'রফিক'}
# করিম ও সালাম বাদ — ওরা দুটোতেই ছিল
```

# 3️⃣ Set Comparison (তুলনা করা)

```python
ছোট = {1, 2, 3}
বড় = {1, 2, 3, 4, 5}
সমান = {1, 2, 3}
আলাদা = {6, 7, 8}

# Subset — ছোট কি বড়-এর ভেতরে আছে?
print(ছোট.issubset(বড়))     # True ✅
print(ছোট <= বড়)             # True ✅

# Superset — বড় কি ছোট-কে ধারণ করে?
print(বড়.issuperset(ছোট))   # True ✅
print(বড় >= ছোট)             # True ✅

# Equal — দুটো কি সমান?
print(ছোট == সমান)           # True ✅

# Disjoint — দুটোতে কি কোনো মিল নেই?
print(ছোট.isdisjoint(আলাদা)) # True ✅
print(ছোট.isdisjoint(বড়))    # False ❌ (মিল আছে)
```

# 4️⃣ Set-এর গুরুত্বপূর্ণ বৈশিষ্ট্য

```python
# ✅ বৈশিষ্ট্য ১ — Duplicate নেই
সংখ্যা = {1, 1, 2, 2, 3, 3}
print(সংখ্যা)   # {1, 2, 3}

# ✅ বৈশিষ্ট্য ২ — Order নেই
s = {5, 3, 1, 4, 2}
print(s)  # যেকোনো order-এ আসতে পারে!
# তাই s[0] করা যাবে না — IndexError!

# ✅ বৈশিষ্ট্য ৩ — শুধু Immutable item রাখা যায়
সঠিক = {1, "Hello", (1, 2)}  # ✅
# ভুল = {1, [1, 2]}  ❌ List রাখা যায় না — unhashable!

# ✅ বৈশিষ্ট্য ৪ — খুব দ্রুত Search!
বিশাল_set = set(range(1000000))
print(999999 in বিশাল_set)  # তাৎক্ষণিক! O(1) ✅

# ✅ বৈশিষ্ট্য ৫ — True ও 1 একই!
s = {1, True, 2, False, 0}
print(s)  # {0, 1, 2} — True=1, False=0 হিসেবে গণনা!
```

# 5️⃣ সবচেয়ে কাজের ব্যবহার — Duplicate বাদ দেওয়া!

```python
# ✅ ব্যবহার ১ — List থেকে Duplicate বাদ
তালিকা = [1, 2, 2, 3, 3, 3, 4, 4, 5]
unique = list(set(তালিকা))
print(unique)  # [1, 2, 3, 4, 5] ✅

# ✅ ব্যবহার ২ — দুটো List-এ কী মিল আছে?
পছন্দ_রহিম = ["ক্রিকেট", "ফুটবল", "সাঁতার"]
পছন্দ_করিম = ["ফুটবল", "দাবা", "সাঁতার"]

মিল = set(পছন্দ_রহিম) & set(পছন্দ_করিম)
print(f"দুজনেরই পছন্দ: {মিল}")
# দুজনেরই পছন্দ: {'ফুটবল', 'সাঁতার'}

# ✅ ব্যবহার ৩ — কোনটা নতুন যোগ হয়েছে?
আগের_ছাত্র = {"রহিম", "করিম", "জলিল"}
এখনকার_ছাত্র = {"রহিম", "করিম", "মামুন", "রফিক"}

নতুন = এখনকার_ছাত্র - আগের_ছাত্র
চলে_গেছে = আগের_ছাত্র - এখনকার_ছাত্র

print(f"নতুন এসেছে: {নতুন}")         # {'মামুন', 'রফিক'}
print(f"চলে গেছে: {চলে_গেছে}")      # {'জলিল'}

# ✅ ব্যবহার ৪ — দ্রুত membership check
নিষিদ্ধ_শব্দ = {"spam", "hack", "virus"}
বার্তা = "this is a hack attempt"

for শব্দ in বার্তা.split():
    if শব্দ in নিষিদ্ধ_শব্দ:  # O(1) — অনেক দ্রুত!
        print(f"সতর্কতা: নিষিদ্ধ শব্দ '{শব্দ}' পাওয়া গেছে!")
```

# 🎯 Interview-এ যা জিজ্ঞেস করতে পারে:

**Q: Set কী এবং এর বৈশিষ্ট্য কী?**
> ✅ Set হলো Unordered, Unindexed এবং Duplicate-মুক্ত collection। এটা internally **Hash Table** ব্যবহার করে তাই search অনেক দ্রুত।

**Q: খালি Set কীভাবে তৈরি করবো?**
> ✅ `set()` দিয়ে। `{}` দিলে Set না, **Dictionary** তৈরি হয়।

**Q: Set-এ কেন Index নেই?**
> ✅ Set **Unordered** — মানে কোনো নির্দিষ্ট ক্রম নেই। তাই `s[0]` করা যায় না।

**Q: Set ও List-এর Search-এ পার্থক্য কী?**
> ✅ List-এ search **O(n)** — প্রতিটা item চেক করতে হয়। Set-এ search **O(1)** — Hash Table-এর কারণে তাৎক্ষণিক। তাই বড় data-তে Set অনেক দ্রুত।

**Q: remove() ও discard() পার্থক্য কী?**
> ✅ `remove()` — item না থাকলে **KeyError** দেয়। `discard()` — item না থাকলে **চুপ থাকে**, কোনো Error নেই। নিরাপদ কোডের জন্য `discard()` ভালো।

**Q: frozenset কী?**
> ✅ frozenset হলো **Immutable Set** — তৈরির পরে পরিবর্তন করা যায় না। Dictionary-তে Key হিসেবে ব্যবহার করা যায়, কিন্তু সাধারণ Set যায় না।
